```python
# 1. feature scaling: technique to standardize the independent features present in the data in a fixed range
# problem: age: 18-20 salary:300000-4000000
# Numbers have very different magnitudes.
# large numbers dominate learning and bias can be caused due to this(Algorithmic Bias)
# --------->1. standardization:Standardization (Z-score scaling) is a technique used to transform your data so that:
# Mean (μ) = 0
# Standard Deviation (σ) = 1
# value - mean/std
# when to use:
# Features have different scales
# When data is not in a fixed range
# k-mean , KNN ,PCA, gradient descent , Artificial neural network
# **** has no effect on outliers
#---------->2. Normalization(MinMaxScaler): Rescales values into a fixed range, usually 0 to 1 without disorting differences in the range of values
#              Basicly retain the data format.
              # properties : after scaling max value : 1 and min value :0
             # Use when:
             # when we already know the minimun and maximum in the given data.
             # Data has known boundaries
             # No extreme outliers
             # Distance-based algorithms
# very sensitive to the outlier if the data set have more outlier values will compress more
            #  Robust Scaling: Uses median and IQR (Interquartile Range) instead of mean & std.
            # use when :Dataset contains many outliers
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/covid_toy - covid_toy.csv')
```

```python
df.isnull().sum()  # handle null values
```

```python
df['fever'] = df['fever'].fillna(df['fever'].mean())
```

```python
df.head(2)
```

```python
df = df.drop(columns= ['city'])
df['gender'] = df['gender'].map({'Female':1,'Male':0 })
df['cough'] = df['cough'].map({'Mild':0,'Strong':1 })
```

```python
x = df.drop(columns = ['has_covid']) ### Input_data or training data
y = df['has_covid'] ### output or testing data
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(
    x,y,test_size=0.2 , random_state= 42
)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
x_train_sc = sc.fit(x_train)
x_train_sc_new = sc.transform(x_train)
x_test_sc_new = sc.transform(x_test)
x_train_sc_df = pd.DataFrame(x_train_sc_new , columns = x_train.columns)
```

Train data → past (learning phase)
Test data → future (evaluation phase)
# New Section
Fit(learn statistics) on train so learning uses only past knowledge; transform(apply) test,train so evaluation simulates future predictions.
*** if i fit on test data then it can cause data leakage
# New Section
We scale X because feature magnitude affects model learning and optimization. We usually don’t scale y because it represents the prediction target and does not influence feature weighting and are independent of each other , except in some regression or deep learning cases where scaling improves optimization stability.

```python
# ENCODING : Encoding in feature engineering means converting categorical (text)
          #  data into numerical form so that machine learning models can understand and use it.
# Why encoding is needed ?
# ---> ML models perform mathematical operations → need numbers
# ---> Encoding helps represent categories in a structured way

# 1. Nominal Data: categories with NO natural order
                  # eg: City → Delhi, Mumbai, Chennai
# 2.Ordinal Data: categories with a meaningful order or ranking
                  # eg: Small < Medium < Large
```

```python
# 1. Label Encoding: Each category gets a unique number and always used on Output column not on input data as it give value implicitly
# 2. Ordinal Encoding : Similar to label encoding but order is meaningful.
```
