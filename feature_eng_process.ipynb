```python
# 1. feature scaling: technique to standardize the independent features present in the data in a fixed range
# problem: age: 18-20 salary:300000-4000000
# Numbers have very different magnitudes.
# large numbers dominate learning and bias can be caused due to this(Algorithmic Bias)
# --------->1. standardization:Standardization (Z-score scaling) is a technique used to transform your data so that:
# Mean (μ) = 0
# Standard Deviation (σ) = 1
# value - mean/std
# when to use:
# Features have different scales
# When data is not in a fixed range
# k-mean , KNN ,PCA, gradient descent , Artificial neural network
# **** has no effect on outliers
#---------->2. Normalization(MinMaxScaler): Rescales values into a fixed range, usually 0 to 1 without disorting differences in the range of values
#              Basicly retain the data format.
              # properties : after scaling max value : 1 and min value :0
             # Use when:
             # when we already know the minimun and maximum in the given data.
             # Data has known boundaries
             # No extreme outliers
             # Distance-based algorithms
# very sensitive to the outlier if the data set have more outlier values will compress more
            #  Robust Scaling: Uses median and IQR (Interquartile Range) instead of mean & std.
            # use when :Dataset contains many outliers
```

```python
# how to handle nominal categorical values
# 1. one-hot encoding: Converts categorical data to numerical format and Creates separate columns for each category
                      #  Uses binary values (0 or 1)
#                        Ensures no ordinal relationship is introduced
# eg: City → Delhi, Mumbai, Chennai
# used for Avoids wrong relationships as ml algo understands in numbers

####### Dummy Variable Trap**********
#:> In One-Hot Encoding, if you include all categories, it causes multicollinearity.
# Multicollinearity is a condition in which two or more independent variables (features) in a input have mathematical relation
# To fix it: Drop one column → drop='first' in encoding
# In models like Linear Regression, it reduces reliability
```

```python
# When One-Hot Encoding (OHE) is applied to a feature with many categories, it creates a large number of columns → this leads to:
# High memory usage
# Slower training
# Risk of overfitting
# To fix it: 1. Keep only the most frequent categories and group the rest.
          #  2. Use Embeddings
          #  3. Use models that handle categories direct
# Get_dummies: not used mostly as No pipeline support and does not handle unseen categories
```

```python
# adding simple imputer to fever col
si = SimpleImputer()
X_train_fever = si.fit_transform(X_train[['fever']])

# also the test data
X_test_fever = si.transform(X_test[['fever']])

X_train_fever.shape
```

```python
oe = OrdinalEncoder(categories=[['Mild','Strong']])
X_train_cough = oe.fit_transform(X_train[['cough']])

# also the test data
X_test_cough = oe.transform(X_test[['cough']])

X_train_cough.shape
```

```python
# OneHotEncoding -> gender,city
ohe = OneHotEncoder(drop='first',sparse_output=False)
X_train_gender_city = ohe.fit_transform(X_train[['gender','city']])

# also the test data
X_test_gender_city = ohe.transform(X_test[['gender','city']])

X_train_gender_city.shape
```

```python
# Extracting Age
X_train_age = X_train.drop(columns=['gender','fever','cough','city']).values

# also the test data
X_test_age = X_test.drop(columns=['gender','fever','cough','city']).values

X_train_age.shape
```

```python
X_train_transformed = np.concatenate((X_train_age,X_train_fever,X_train_gender_city,X_train_cough),axis=1)
X_train_transformed.shape
```

Mentos Zindagi (using Column transformner)

```python
transformer = ColumnTransformer(transformers=[
    ('t1',SimpleImputer() ,['fever']),
    ('t2',OrdinalEncoder(categories=[['Mild','Strong']]),['cough']),
    ('t3',OneHotEncoder(sparse_output=False ,drop ='first') , ['gender','city']),
],remainder='passthrough')
```
